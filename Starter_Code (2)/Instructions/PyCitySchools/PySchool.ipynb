```python
#Importing Dependencies and Setup
import pandas as pd
import numpy as np

#Loading Resource files
school_load = "Resources/schools_complete.csv"
student_load = "Resources/students_complete.csv"

#Read csvs into Pandas DataFromes
school_data = pd.read_csv(school_load)
student_data = pd.read_csv(student_load)

#Combine both data sheets into one
combined_data = pd.merge(school_data, student_data, how="left", on=["school_name", "school_name"])
combined_data.head()
```

```python
#Determine how many schools
school_total = len(school_data["school_name"].unique())
school_total
```

```python
#Determine total number of students 
student_total = int(student_data["Student ID"].nunique())
student_total
```

```python
#Determine the total budget sum of all schools
budget_total = school_data["budget"].sum()
budget_total
```

```python
#Calculate the average math score among all schools
avg_math = student_data["math_score"].mean()
avg_math
```

```python
#Calculate the average reading scores among all schools
avg_read = student_data["reading_score"].mean()
avg_read
```

```python
#Calculate the percentage of students passing math among all schools
per_passing_math = ((student_data["math_score"]>70).sum() / student_data["math_score"].count())*100
per_passing_math
```

```python
#Calculate the percentage of students passing reading among all schools
per_passing_read = ((student_data["reading_score"]>70).sum() / student_data["reading_score"].count())*100
per_passing_read
```

```python
#Calculate the percentage of students passing math and reading among all schools
ovr_passing = (per_passing_math + per_passing_read)/2
ovr_passing
```

```python
#Creating a data frame to display the above information
info_table = pd.DataFrame({"Total Schools":[school_total], "Total Students":[student_total], "Total Budget":[budget_total], 
                           "Average Math Score":[avg_math], "Average Reading Score":[avg_read], "% Passing Math":[per_passing_math], 
                           "% Passing Reading":[per_passing_read], "% Overall Passing":[ovr_passing]})
info_table
```

```python
#Identify the type of each school
school_type = school_data.set_index(["school_name"])["type"]
school_type
```

```python
#Determine the total number of students in each school
student_total_per_school = combined_data["school_name"].value_counts()
student_total_per_school
```

```python
#Calculate each school's total budget
budget_per_school = combined_data.groupby(["school_name"]).mean()["budget"]
budget_per_school
```

```python
#Calculate each school's budget for the individual student
budget_per_student = budget_per_school/student_total_per_school
budget_per_student
```

```python
#Calculate each school's average math grade
math_avg_per_school = combined_data.groupby(["school_name"]).mean()["math_score"]
math_avg_per_school
```

```python
#Calculate each school's average reading grade
read_avg_per_school = combined_data.groupby(["school_name"]).mean()["reading_score"]
read_avg_per_school
```

```python
#Calculate the percentage of students passing math for each school
passing_math = combined_data[(combined_data["math_score"]>70)]
passing_math_per_school = passing_math.groupby(["school_name"]).count()["student_name"]
perc_pass_math_per_school = (passing_math_per_school/student_total_per_school)*100
perc_pass_math_per_school
```

```python
#Calculate the percentage of students passing reading for each school
passing_reading = combined_data[combined_data["reading_score"]>70]
passing_reading_per_school = passing_reading.groupby(["school_name"]).count()["student_name"]
perc_pass_read_per_school = (passing_reading_per_school/student_total_per_school)*100
perc_pass_read_per_school
```

```python
#Calculate the percentage of students passing both math and reading for each school
ovr_pass_per_school = (perc_pass_math_per_school + perc_pass_read_per_school)/2
ovr_pass_per_school
```

```python
#Create a data frame to list out above information
school_sum_df = pd.DataFrame({"School Type":school_type, "Total Students":student_total_per_school, 
                              "Total School Budget ($)":budget_per_school, "Per Student Budget ($)":budget_per_student, 
                              "Average Math Score":math_avg_per_school, "Average Reading Score":read_avg_per_school, 
                              "% Passing Math":perc_pass_math_per_school, "% Passing Reading": perc_pass_read_per_school, 
                              "% Overall Passing": ovr_pass_per_school})
school_sum_df
```

```python
#Data frame to display the top 5 performing schools based on the percentage of students passing math and reading
top_schools = school_sum_df.sort_values("% Overall Passing", ascending=False)
top_schools.head()
```

```python
#Data frame to display the bottom 5 performing schools based on the percentage of students passing math and reading
bot_schools = school_sum_df.sort_values("% Overall Passing")
bot_schools.head()
```

```python
#Data frame to display the average math scores per grade per school
math_9 = student_data.loc[student_data["grade"] == "9th"].groupby(["school_name"])["math_score"].mean()
math_10 = student_data.loc[student_data["grade"] == "10th"].groupby(["school_name"])["math_score"].mean()
math_11 = student_data.loc[student_data["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
math_12 = student_data.loc[student_data["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()
math_by_grade = pd.DataFrame({"9th":math_9, "10th":math_10, "11th":math_11, "12th":math_12})
math_by_grade
```

```python
#Data frame to display the average reading scores per grade per school
read_9 = student_data.loc[student_data["grade"] == "9th"].groupby(["school_name"])["reading_score"].mean()
read_10 = student_data.loc[student_data["grade"] == "10th"].groupby(["school_name"])["reading_score"].mean()
read_11 = student_data.loc[student_data["grade"] == "11th"].groupby(["school_name"])["reading_score"].mean()
read_12 = student_data.loc[student_data["grade"] == "12th"].groupby(["school_name"])["reading_score"].mean()
read_by_grade = pd.DataFrame({"9th":read_9, "10th":read_10, "11th":read_11, "12th":read_12})
read_by_grade
```

```python
#Assign each school with corresponding spending range per student
score_bins = [0, 585, 630, 645, 680]
score_names = ["<$585", "$585-630", "$630-645", "$645-680"]

school_sum_df["Spending Ranges (Per Student)"] = pd.cut(budget_per_student, score_bins, labels = score_names)
school_scores = school_sum_df.drop(columns=["Total Students", "Total School Budget ($)", "School Type", "Per Student Budget ($)"])
school_scores
```

```python
#Table that displays the spending range per student and the corresponding academic scores
school_spending = school_scores.groupby(["Spending Ranges (Per Student)"])
school_spending.mean()
```

```python
#Assign each school with its corresponding school size
size_bins = [0, 1000, 2000, 5000]
size_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_sum_df["School Size"] = pd.cut(student_total_per_school, size_bins, labels = size_names)
school_size = school_sum_df.drop(columns=["Total Students", "Total School Budget ($)", "School Type", "Per Student Budget ($)"])
school_size
```

```python
#Table that displays the total school size and the corresponding academic scores
scores_size = school_size.groupby(["School Size"])
scores_size.mean()
```

```python
#Assign each category with its corresponding school type
math_score_type = school_sum_df.groupby(["School Type"]).mean()["Average Math Score"]
read_score_type = school_sum_df.groupby(["School Type"]).mean()["Average Reading Score"]
perc_math_type = school_sum_df.groupby(["School Type"]).mean()["% Passing Math"]
perc_read_type = school_sum_df.groupby(["School Type"]).mean()["% Passing Reading"]
ovr_pass_type = school_sum_df.groupby(["School Type"]).mean()["% Overall Passing"]

##Table that displays the type of school and the corresponding academic scores
score_type = pd.DataFrame({"Average Math Score":math_score_type, 
                           "Average Reading Score":read_score_type, 
                           "% Passing Math":perc_math_type, 
                           "% Passing Reading":perc_read_type, 
                           "% Overall Passing":ovr_pass_type})
score_type
```
